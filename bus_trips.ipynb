```python
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
import matplotlib.pyplot as plt

#df = pd.read_csv('C:\Users\Chintan\Documents\Desktop\\bus_trips.csv')
df =pd.read_csv("https://docs.google.com/spreadsheets/d/1vDKGuI8bGjtj0D-GGv5zremg7oxfoFS-hxukRuBpqoQ/pub?output=csv")
df = df.drop('Unnamed: 3', axis=1)
df

def f(x):
    if x == x:
        y = x.split(':')
        if len(y) == 2:
            return int(y[0])*60 + int(y[1])
        else:
            return 0
    else:
        return 0
df['Time'] = df['End'].apply(f) - df['Start'].apply(f)

def removeInvalid(x):
    if x<=0 or x>100:
        return 0
    else:
        return x

def removeNegsOrZeroes(x):
    if x <=0 or x > 100:
        return nan
    else:
        return x
    
df['Time'] = df['Time'].apply(removeInvalid)
df = df[df['Time'] !=0]
#df.plot(x='Date',y = 'Time')
```

```python
df['DateConverted'] = pd.to_datetime(df['Date'].astype(str), format='%m/%d/%Y')
```

```python
def makeBuckets(x):
    y = x.split(':')
    hrint = int(y[0])
    minint = int(y[1])
    if minint <15:
        minint = 0
    elif minint <30:
        minint = 15
    elif minint <45:
        minint = 30
    elif minint <60:
        minint = 45
    
    #minfinal = minint - 15
    #if minfinal <0:
     #   minfinal = 60 + minfinal
      #  hrint = hrint -1
    return str(hrint) + ":" + str(minint)

        
```
